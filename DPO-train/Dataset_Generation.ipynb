```python
import json

# List of .txt file paths
file_paths = [
    "file_singleresponses_0to2k.txt",
    "file_singleresponses_2kto4k.txt",
    "file_singleresponses_4kto6k.txt",
    "file_singleresponses_6kto8k.txt",
    "file_singleresponses_8kto10k.txt"
]

# Initialize a list to hold all parsed JSON objects
json_objects = []

# Process each file in the list
for file_path in file_paths:
    try:
        # Open the file and read it line by line
        with open(file_path, 'r') as file:
            for line in file:
                # Strip leading/trailing whitespace and parse as JSON
                try:
                    json_object = json.loads(line.strip())
                    json_objects.append(json_object)
                except json.JSONDecodeError:
                    print(f"Skipping invalid JSON line in file: {file_path}")

    except FileNotFoundError:
        print(f"File not found: {file_path}")

# Output the results
print(f"Parsed a total of {len(json_objects)} JSON objects.")
for obj in json_objects[:2]:  # Display the first two objects for brevity
    print(json.dumps(obj, indent=4))  # Pretty-print JSON

# Optional: Save combined JSON objects to a new file
output_file = "combined_responses.json"
with open(output_file, "w") as file:
    json.dump(json_objects, file, indent=4)

print(f"Combined JSON objects saved to {output_file}")
```

```python
# List of JSON file paths
file_paths = [
    "llavadataset0to2k.json",
    "llavadataset2kto4k.json",
    "llavadataset4kto6k.json",
    "llavadataset6kto8k.json",
    "llavadataset8kto10k.json"
]

# Combined data storage
combined_data = []

# Loop through each file path, open, and load data
for file_path in file_paths:
    try:
        with open(file_path, "r") as file:
            data = json.load(file)
            combined_data.extend(data)  # Assuming each file contains a list of entries
    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except json.JSONDecodeError:
        print(f"Error decoding JSON in file: {file_path}")

# Check the total number of entries
print(f"Total number of entries: {len(combined_data)}")

# Optional: Save combined data to a new file
output_file = "combined_dataset.json"
with open(output_file, "w") as file:
    json.dump(combined_data, file, indent=4)

print(f"Combined dataset saved to {output_file}")
```

```python
overall_scores = []
for idx, obj in enumerate(json_objects):
    try:
        # Access the nested content
        content = obj['response']['body']['choices'][0]['message']['content']
        # Parse the JSON content
        parsed_content = json.loads(content)
        # Get the overall_score
        overall_score = parsed_content.get('overall_score', None)
        overall_scores.append((idx, overall_score))
    except (KeyError, IndexError, json.JSONDecodeError) as e:
        print(f"Error processing object at index {idx}: {e}")
        overall_scores.append((idx, 3)) # handle errors
```

```python
# Initialize result lists
highest_indices = []
lowest_indices = []

# Process groups of eight
for i in range(0, len(overall_scores), 8):
    # Get the group of eight
    group = overall_scores[i:i+8]

    # Find the highest score and its index
    highest = max(group, key=lambda x: x[1])  # x[1] is the score
    highest_indices.append(highest[0])  # Append the index

    # Find the lowest score and its index
    lowest = min(group, key=lambda x: x[1])
    lowest_indices.append(lowest[0])  # Append the index
```

```python
from datasets import load_dataset

ds = load_dataset("HuggingFaceM4/A-OKVQA")
ds = ds["train"]
from itertools import chain
indices = chain(range(0, 2000), range(2001, 10001))
ten_k = ds.select(list(indices))
```

```python
highest_overall_scores = [overall_scores[idx] for idx in highest_indices]
lowest_overall_scores = [overall_scores[idx] for idx in lowest_indices]
highest_overall_scores[:5]
```

```python
full_responses = [combined_data[idx]['full_response'] for idx in range(len(combined_data))]
flattened_responses = [response for sublist in full_responses for response in sublist]
```

```python
best_responses = [(flattened_responses[item[0]],item[1]) for item in highest_overall_scores]
worst_responses = [(flattened_responses[item[0]],item[1]) for item in lowest_overall_scores]
```

```python
worst_response_texts = [resp[0] for resp in worst_responses]
worst_response_scores = [resp[1] for resp in worst_responses]
best_response_texts = [resp[0] for resp in best_responses]
best_response_scores = [resp[1] for resp in best_responses]

# Create a function to map the worst_response and worst_response_score to each item in the Dataset
def add_worst_response(example, idx):
    # Safely assign worst response and score based on the index
    if idx < len(worst_response_texts):
        example['worst_response'] = worst_response_texts[idx]
        example['worst_response_score'] = worst_response_scores[idx]
        example['best_response'] = best_response_texts[idx]
        example['best_response_score'] = best_response_scores[idx]
    return example

# Apply the function using map to add the new columns
updated_ten_k = ten_k.map(add_worst_response, with_indices=True)
```

```python
def process_responses(example):
    try:
        # Extract "question" from best_full_response

        start = example["best_response"].find("USER:") + len("USER:  \n")
        end = example["best_response"].find("ASSISTANT")
        example["question_full"] = example["best_response"][start:end].strip()


        # Parse down best_full_response to everything after "ASSISTANT"

        best_start = example["best_response"].find("ASSISTANT:") + len("ASSISTANT:")
        example["best_response"] = example["best_response"][best_start:].strip()

        # Parse down worst_full_response to everything after "ASSISTANT"

        worst_start = example["worst_response"].find("ASSISTANT:") + len("ASSISTANT:")
        example["worst_response"] = example["worst_response"][worst_start:].strip()

    except Exception as e:
        print(f"Error processing row: {e}")
        example["question_full"] = None
        example["best_response"] = None
        example["worst_response"] = None
    return example

# Apply the processing function to the dataset
final = updated_ten_k.map(process_responses)
```
